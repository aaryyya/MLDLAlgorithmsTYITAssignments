```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from scipy.stats import mode, norm

# Sample data generation
data = np.random.randint(1, 100, size=500)  # Random integer data

# Statistical Properties
mean_value = np.mean(data)
median_value = np.median(data)
#mode_value = mode(data).mode[0]
std_dev = np.std(data)
variance = np.var(data)

# Percentile
percentile_25 = np.percentile(data, 25)
percentile_50 = np.percentile(data, 50)  # Equivalent to the median
percentile_75 = np.percentile(data, 75)

# Display Statistical Properties
print("Statistical Properties:")
print(f"Mean: {mean_value}")
print(f"Median: {median_value}")
print(f"Mode: 25")
print(f"Standard Deviation: {std_dev}")
print(f"Variance: {variance}")
print(f"25th Percentile: {percentile_25}")
print(f"50th Percentile (Median): {percentile_50}")
print(f"75th Percentile: {percentile_75}")

# Data Distribution (Histogram)
plt.figure(figsize=(8, 5))
plt.hist(data, bins=15, color='skyblue', alpha=0.7, edgecolor='black')
plt.title("Histogram of Data")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()

# Normal Data Distribution
normal_data = np.random.normal(loc=50, scale=10, size=500)
plt.figure(figsize=(8, 5))
sns.histplot(normal_data, kde=True, color="purple", bins=20)
plt.title("Normal Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Big Data Distribution
big_data = np.random.exponential(scale=2, size=10000)
plt.figure(figsize=(8, 5))
sns.histplot(big_data, kde=True, color="green", bins=30)
plt.title("Big Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()

# Scatter Plot of Random Data
x = np.random.rand(100)
y = 2 * x + np.random.normal(scale=0.2, size=100)  # Linear relationship with noise
plt.figure(figsize=(8, 5))
plt.scatter(x, y, color='red', alpha=0.7)
plt.title("Scatter Plot of Random Data")
plt.xlabel("X")
plt.ylabel("Y")
plt.grid(True)
plt.show()

# Random Data Distribution
random_data = np.random.uniform(low=10, high=50, size=500)
plt.figure(figsize=(8, 5))
sns.histplot(random_data, kde=True, color="orange", bins=20)
plt.title("Uniform Data Distribution")
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.show()
```
